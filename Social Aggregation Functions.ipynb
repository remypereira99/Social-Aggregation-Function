```python
import random
```

```python
class AdmissProf:
    
    """
    Class of admissable voter profiles
    """
    
    def __init__(self,v1,v2,v3):
        self.v1 = v1
        self.v2 = v2
        self.v3 = v3
        self.voters = [v1,v2,v3]
        
    def maj_rule(self):
        
        social_choice = 'x'
        
        if (self.v1 + self.v2 + self.v3)>0:
            social_choice = 1
            print(f'The social choice is {social_choice}')
        if (self.v1 + self.v2 + self.v3)<0:
            social_choice = -1
            print(f'The social choice is {social_choice}')
        if (self.v1 + self.v2 + self.v3)==0:
            social_choice = 0
            print(f'The social choice is {social_choice}')
    
    def unan_rule(self):
        
        social_choice = 'x'
        
        if (self.v1 + self.v2 + self.v3)==3:
            social_choice = 1
            print(f'The social choice is {social_choice}')
        elif (self.v1 + self.v2 + self.v3)==-3:
            social_choice = -1
            print(f'The social choice is {social_choice}')
        elif (self.v1 + self.v2 + self.v3)!=(3 or -3):
            social_choice = 0
            print(f'The social choice is {social_choice}')
        
    def dictatorship(self):
        social_choice = random.choice(self.voters)
        print(f'The social choice is {social_choice}')
```

```python
p1 = AdmissProf(1,1,1)
p2 = AdmissProf(1,1,-1)
p3 = AdmissProf(1,-1,1)
p4 = AdmissProf(-1,1,1)
p5 = AdmissProf(-1,-1,1)
p6 = AdmissProf(-1,1,-1)
p7 = AdmissProf(1,-1,-1)
p8 = AdmissProf(-1,-1,-1)
```
